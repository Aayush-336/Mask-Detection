```python
haar_data = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
```

```python
# rect_arr = haar_data.detectMultiScale(img)
```

```python
data = []
for category in categories:
    path = os.path.join(directory,category)
    data.clear()
    c = 0
    for img in os.listdir(path):
        img_path = os.path.join(path,img)
#         print(img_path)
        image = cv2.imread(img_path)
#         rect_arr = haar_data.detectMultiScale(image)
#         for x,y,w,h in rect_arr:
#             face = image[y:y+h,x:x+w,:]
#             face = cv2.resize(face,(50,50))
        image = cv2.resize(image,(100,100))
        data.append(image)
        
        c+=1
    
    print(c)
    np.save(category+'1'+'.npy',data)
```

```python
data2 = []
cap = cv2.VideoCapture(0)
while True:
    flag,img = cap.read()
    if flag:
        rect_arr = haar_data.detectMultiScale(img)        
        for x,y,w,h in rect_arr:
            cv2.rectangle(img,(x,y),(x+w,y+w),(255,0,255),4)
            face = img[y:y+h,x:x+w,:]
            face = cv2.resize(face,(100,100))
            data2.append(face)
#         print(len(data))
        cv2.imshow('result',img)
        if cv2.waitKey(2) == 27 or len(data2)>100:
            break

cap.release()
cv2.destroyAllWindows()
data.extend(data2)
print(len(data))
```

```python
np.save(category+'1'+'.npy',data)
```
